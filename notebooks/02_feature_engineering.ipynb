```python
import pandas as pd

# ---------------------------------------------------------
# ADIM 1: Verileri Yükleme
# ---------------------------------------------------------
train_df = pd.read_csv("../data/raw/train.csv", index_col='id')
test_df = pd.read_csv("../data/raw/test.csv", index_col='id')

print("--- 1. Hedef Değişkeni Sayısallaştırma ---")
# Hedef değişkeni ayırıp 1 ve 0'a çeviriyoruz (Yes=1, No=0)
y_train = train_df['Churn'].map({'Yes': 1, 'No': 0})
train_df.drop('Churn', axis=1, inplace=True)

print("--- 2. Kategorik Değişkenleri Sayısallaştırma (Encoding) ---")
# PROFESYONEL İPUCU: Train ve Test setine aynı kodlamayı uygulamak için 
# verileri geçici olarak birleştirip işlem yapıyor, sonra tekrar ayırıyoruz.
combined_df = pd.concat([train_df, test_df], keys=['train', 'test'])

# Olası TotalCharges hatasını düzeltme (Boşlukları sıfır yapıyoruz)
combined_df['TotalCharges'] = pd.to_numeric(combined_df['TotalCharges'], errors='coerce').fillna(0)

# Kategorik değişkenleri tespit edip One-Hot Encoding yapıyoruz (drop_first=True ile gereksiz kolonları atıyoruz)
categorical_cols = combined_df.select_dtypes(include=['object']).columns
combined_df_encoded = pd.get_dummies(combined_df, columns=categorical_cols, drop_first=True)

# İşlem bitince verileri tekrar Train ve Test olarak ayırıyoruz
X_train = combined_df_encoded.xs('train')
X_test = combined_df_encoded.xs('test')

print("İşlem tamamlandı!")
print(f"Yeni Eğitim Seti Boyutu (X_train): {X_train.shape}")
print(f"Yeni Test Seti Boyutu (X_test): {X_test.shape}")
```

```python
# ---------------------------------------------------------
# ADIM 3: İşlenmiş Verileri Diske Kaydetme
# ---------------------------------------------------------
print("--- 3. İşlenmiş Verileri Kaydetme ---")

X_train.to_csv("../data/processed/X_train_processed.csv")
y_train.to_csv("../data/processed/y_train_processed.csv")
X_test.to_csv("../data/processed/X_test_processed.csv")

print("✅ Temizlenmiş ve sayısallaştırılmış veriler 'data/processed' klasörüne başarıyla kaydedildi!")
```
